```python
#データベースとの接続について
```

```python
import sqlite3
```

```python
df_csv = pd.read_csv('data.csv',encoding='shift-jis')
df_csv
```

```python
dbname = 'TEST.db'#作成するデータベースの名前
conn = sqlite3.connect(dbname)#コネクションオブジェクトの作成
conn.close()#データベースの切断
```

```python
conn = sqlite3.connect(dbname)
df_csv.to_sql('test_table',conn,if_exists = 'replace')#if_existsは、同じテーブル名が存在する場合の動作replaceは、テーブルを削除してから新規で作成
conn.close()
```

```python
#テーブルからデータの取得
```

```python
conn = sqlite3.connect(dbname)
df_db = pd.read_sql('SELECT * FROM test_table' ,conn)
conn.close()
```
