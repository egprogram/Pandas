```python
#結合について
```

```python
import pandas as pd
```

```python
df01 = pd.DataFrame({'氏名':['高橋','伊藤','鈴木','佐藤'],
                    'クラス':['a001','a001','a001','a001'],
                    '数学':[1,2,3,4]})
df01
```

```python
df02 = pd.DataFrame({'氏名':['高橋','伊藤','鈴木','佐藤'],
                    'クラス':['a001','a001','a001','a001'],
                    '国語':[5,6,7,8]})
df02
```

```python
pd.merge(df01,df02, on='氏名')#on=結合キー
```

```python
pd.merge(df01,df02, on='氏名',suffixes=['_left','_right'])
```

```python
pd.merge(df01,df02, on='氏名',suffixes=['','_重複'])
```

```python
df02_name = pd.DataFrame({'名前':['高橋','伊藤','鈴木','佐藤'],
                    'クラス':['a001','a001','a001','a001'],
                    '国語':[5,6,7,8]})
df02_name
```

```python
#カラム名が違うもの同士の結合について
```

```python
pd.merge(df01,df02_name,left_on='氏名',right_on='名前')
```

```python
df03 = pd.DataFrame({'名前':['高橋','高橋'],
                     'クラス':['a001','a002'],
                     '英語':[15,16]})
df03
```

```python
pd.merge(df01,df03,left_on=['氏名','クラス'],right_on=['名前','クラス'])
```

```python
#結合方法innner,right,left,right,outer
```

```python
df04 = pd.DataFrame({'氏名':['高橋','伊藤','渡辺','加藤'],
                     'クラス':['a001','a001','a001','a001'],
                     '国語':[5,6,7,8]})
df04
```

```python
pd.merge(df01,df04,on='氏名',how='left')#how~結合方法
```

```python
pd.merge(df01,df04,on='氏名',how='right')
```

```python
#ちなみにデフォルトはinner
```

```python
pd.merge(df01,df04,on='氏名')
```

```python
df05 = pd.DataFrame({'クラス':['a001','a001','a001','a001'],
                     '数学':[5,6,7,8]}
                     ,index=['高橋','伊藤','鈴木','佐藤'])
df05
```

```python
df02
```

```python
pd.merge(df05,df02,left_index=True,right_on='氏名')
```

```python
pd.merge(df05,df02,left_index=True,right_on='氏名',sort=True)
```
